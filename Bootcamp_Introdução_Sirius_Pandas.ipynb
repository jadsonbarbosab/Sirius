```python
df1 = pd.DataFrame(
    {
        "A" : ["A0","A1","A2","A3"],
        "B" : ["B0","B1","B2","B3"],
        "C" : ["C0","C1","C2","C3"],
        "D" : ["D0","D1","D2","D3"],
    },
    index = [0,1,2,3],
)

df2 = pd.DataFrame(
    {
        "A" : ["A4","A5","A6","A7"],
        "B" : ["B4","B5","B6","B7"],
        "C" : ["C4","C5","C6","C7"],
        "D" : ["D4","D5","D6","D7"],
    },
    index = [4,5,6,7],
)
```

```python
df1.head()
```

```python
df2.head()
```

```python
result = pd.concat([df1,df2]) #COMBINANDO 2 DATAFRAMES
result.head()
```

```python
df4 = pd.DataFrame(
    {
        "B" : ["B2","B3","B6","B7"],
        "D" : ["D2","D3","D6","D7"],
        "F" : ["F2","F3","F6","F7"],
    },
    index = [4,5,6,7],
)
```

```python
df4.head()
```

```python
result_2 = pd.concat([df1,df4], axis =1, join = 'outer') #COMBINANDO 2 DATAFRAMES
result_2.head()
```

```python
result_2 = pd.concat([df1,df4], axis =1, join = 'inner') #COMBINANDO 2 DATAFRAMES
result_2.head()
```
