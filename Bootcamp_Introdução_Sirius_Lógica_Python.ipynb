### **WHILE**


```python
#Programa só para quando usurário digitar 0; Fazer soma de todosos digitados
n = 1
soma = 0
while n!= 0:
  num = int(input('Digite um valor: '))
  soma = soma + num

  if num == 0:
    n = 0
print(f'A soma total dos números digitados foi {soma}')
```

Exercício 1 (Aquecimento): Vamos criar uma calculadora (simples) com as operações soma, subtração, multiplicação, divisão.
Dica: crie funções para tarefas específicas como, por exemplo, uma função soma(a,b), uma função subtracao(a,b), ... e depois use estas funções para criar a função calculadora().

Lembre-se de que não se pode dividir por zero.

```python
def soma(a,b):
  print(a+b)

def sub(a,b):
  print(a-b)

def mult(a,b):
  print(a*b)

def div(a,b):
  if b == 0:
    print('Não se pode dividir por 0.')
  else:
    print(a/b)
```

```python
def calculadora():
  i = 0
  while i == 0:
    num = int(input('''Digite a sua opção:
    1) Soma
    2) Subtração
    3) Divisão
    4) Multiplicação
    5) Parar calculadora \n'''))

    if num == 5:
      break

    a = int(input('Digita o primeiro valor: '))
    b = int(input('Digita o segundo valor: '))

    if num == 1:
      print('A soma dos números é ')
      soma(a,b)
    
    elif num == 2:
      print('A subtração dos números é ')
      sub(a,b)
      
    
    elif num == 3:
       print('A divisão dos números é ')
       div(a,b)
     
    
    elif num == 4:
      print('A multiplicação dos números é ')
      mult(a,b)
      
    
    else:
      print('Digite uma opção válida')
  print('Calculadora encerrada')
```

Exercício 2: Em um curso da Sirius temos alunos dos mais variados perfis, tendo diferentes faixas etárias, morando em diferentes estados do Brasil e que tiraram diferentes notas. Essas informações estão em diferentes listas, uma somente com os nomes, outras somente com as notas, outra com a idades e outro com os estados de residência.
Com base nisso crie funções que ajudem a responder as perguntas considerando os dados a seguir.

```python
# dados
nomes = ['João', 'José', 'Fernando', 'Rafaela', 'Kleber', 'Laercio', 'Carlos','Paula','Adriana','Walkyria','Guilherme','Gustavo','Gabriel', 'Mayara', 'Ana', 'Daniela', 'Rafael', 'Daiane', 'Maria', 'Mateus', 'Monique', 'Brian', 'Henrique', 'Alberto', 'Mônica', 'Bruce', 'Simone']
notas = [8,7,6,9,3,8,7,5,9,1,4,5,8,4,9,7,5,6,2,7,4,5,8,7,9,10,5]
idades = [25,17,56,32,27,31,48,24,41,50,29,22,30,31,28,52,27,33,35,36,38,26,29,35,38,41,46]
estados = ['PR','SP','SP','MG','RJ','MG','PB','AM','SC','MS','SP','MG','AM','RO','MG','BA','MG','ES','CE','MG','PA','PI','PE','MG','RR','MG','SE']
```

```python
#1) Crie uma função que seleciona o nome de todas as pessoas que residem em um determinado estado. Por exemplo, 
#a função alunos_por_estado(nomes, estados, 'SP') deve retornar todos os nomes dos alunos que moram em São Paulo.
```

```python
def alunos_por_estado(lista_nomes, lista_estados, estado):
  nomes = []
  for i in range(len(lista_estados)):
    if lista_estados[i] == estado:
      nomes.append(lista_nomes[i])
  return nomes

alunos_por_estado(nomes, estados, 'MG')
```

```python
#2) Crie uma função que calcula quantas pessoas foram aprovadas por estado, considerando média 6. Por exemplo, 
#a função quant_aprov_estado(estados, notas, 'SP') deve retornar quantos alunos foram aprovados no estado de São Paulo.
```

```python
def quant_aprov_estado(lista_estados, lista_notas, estado):
  count = 0
  for i in range(len(lista_estados)):
    if lista_estados[i] == estado:
      if notas[i] >=6:
        count = count + 1
  print(f'A quantidade de alunos de {estado} com nota acima de 6 é {count}')

quant_aprov_estado(estados, notas, 'MG')
```

```python
#3) Crie uma função que calcula quantos alunos há por estado. Por exemplo, a função alunos_por_estado(estados) deve retornar uma mensagem dizendo quantos alunos há em cada estado.

def alunos_por_estado(lista_estados):
  estados_sem_repetidos = list(set(lista_estados))
  for estado in estados_sem_repetidos:
    count = 0
    for i in range(len(lista_estados)):
      if lista_estados[i] == estado:
        count = count + 1
    print(f'A quantidade de alunos de {estado} é {count}')
```

```python
alunos_por_estado(estados)
```

```python
#Crie uma função que calcula a média das notas dos alunos de um determinado estado. Por exemplo, a função media_por_estado(notas, estados, 'SP') 
#deve retornar a média das notas do alunos (com duas casas decimais) do estado de São Paulo.

def media_por_estado(lista_notas, lista_estados, estado):
    count = 0
    soma = 0
    for i in range(len(lista_estados)):
      if lista_estados[i] == estado:
        soma = soma + lista_notas[i]
        count = count + 1
    media = soma / count
    print(soma,count)
    print(f'A média dos alunos do {estado} é {media:.2f}')

media_por_estado(notas, estados, 'SP')
```

```python
#5) Crie uma função que calcula a média das idades dos alunos aprovados (considerando media 6) de um determinado estado. 
#Por exemplo, a função media_idade_aprov_por_estado(notas, estados, idades, 'SP') deve retornar a média das idades do alunos (com duas casas decimais) 
#aprovados que residem no estado de São Paulo.
def media_idade_aprov_por_estado(lista_notas, lista_estados, lista_idades, estado):
    #count = 0
    soma = 0
    len_idade = []
    for i in range(len(lista_estados)):
      if (lista_estados[i] == estado and lista_notas[i] >=6):
        
          soma = soma + lista_idades[i]
          #count += 1
          len_idade.append(lista_idades[i])
    media = soma / len(len_idade)
    print(soma,len(len_idade))
    print(f'A média dos alunos do {estado} é {media:.2f}')

media_idade_aprov_por_estado(notas, estados, idades, 'AM')
```
